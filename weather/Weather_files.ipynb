```python
meteo_df=pd.read_csv('TorrePellice_from_2021-01-01_to_2021-12-31',index_col='Date_Time',parse_dates=[0,])
meteo_df.head()
```

```python
meteo=meteo_df.resample('H').mean().interpolate()
meteo.head()
```

```python
meteo.columns
```

```python
latitude = 45.0619883  # latitude  N
longitude = 7.6602814  # longitude E
elevation = 250  # meters above the sea
```

```python
rad=pd.DataFrame(meteo['Solar_W/m2'].values,index=meteo.index,columns=['ghi'])
```

```python
rad=dec.main(rad,latitude,longitude,elevation)
```
