```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = []
dfs.append(pd.read_csv("../cache/results/ach0.csv"))
dfs.append(pd.read_csv("../cache/results/ach2.csv"))
dfs.append(pd.read_csv("../cache/results/ach4.csv"))
dfs.append(pd.read_csv("../cache/results/ach6.csv"))
```

```python
temperature = []
for i in range(4):
    avg_temp= dfs[i]['MAINXGROUND:ZONE1:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE5:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE4:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE6:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE10:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE3:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE2:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE7:Zone Mean Air Temperature [C](TimeStep:ON)']
    avg_temp+=dfs[i]['MAINXGROUND:ZONE8:Zone Mean Air Temperature [C](TimeStep:ON)']
    
    temperature.append(avg_temp/9.0)
```

```python
plt.figure(figsize=(30, 12), dpi=80)
plt.plot(temperature[0],color="r",label="ach0")
plt.plot(temperature[1],color="g",label="ach2")
plt.plot(temperature[2],color="b",label="ach4")
plt.plot(temperature[3],color="c",label="ach6")
plt.legend()
```
